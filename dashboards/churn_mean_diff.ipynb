```python
import pandas as pd
import numpy as np
df = pd.read_csv('../data/ecommerce_churn_data.csv')

# Fill missing with KNN as in notebook
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=5)
cols_to_impute = ['Tenure', 'WarehouseToHome', 'HourSpendOnApp',
                  'OrderAmountHikeFromlastYear', 'CouponUsed',
                  'OrderCount', 'DaySinceLastOrder']
df[cols_to_impute] = imputer.fit_transform(df[cols_to_impute])

# Numeric columns
numeric_cols = df.select_dtypes(include='number').columns.drop('Churn')
print('Numeric variables analyzed:', list(numeric_cols))

# Mean by Churn
means = df.groupby('Churn')[numeric_cols].mean()
print('\nMeans by Churn:')
print(means)

# Differences
diffs = means.iloc[1] - means.iloc[0]
print('\nMean differences (Churn1 - Churn0):')
print(diffs.sort_values(key=abs, ascending=False))

# Significant differences (e.g., |diff| > 0.5 std or threshold based on boxplot variation)
significant = diffs[diffs.abs() > diffs.std() * 0.5]
print('\nVariables with significant mean difference based on boxplot variation:')
print(significant.index.tolist())
```
